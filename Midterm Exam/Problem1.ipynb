```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Define the Transfer Function G(s) = (s + 1) / (s^4 + 4s^3 + 7s^2 + 6s + 3)
num = [1, 1]
den = [1, 4, 7, 6, 3]
sys = ct.TransferFunction(num, den)
```

```python
print("System Transfer Function:")
print(sys)
```

```python
# 2. Performance Metrics
# Calculate step response and metrics
t, y = ct.step_response(sys)
info = ct.step_info(sys)
```

```python
print("\n--- Transient Analysis ---")
print(f"Rise Time (10-90%): {info['RiseTime']:.4f} s")
print(f"Settling Time (2%): {info['SettlingTime']:.4f} s")
print(f"Overshoot:          {info['Overshoot']:.4f}%")
print(f"Steady-state Value: {y[-1]:.4f}")
```

```python
# 3. Plotting
# Plot Step Response
plt.plot(t, y, label='Step Response', color='blue', linewidth=2)
plt.axhline(y[-1], color='red', linestyle='--', label='Steady State')
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
```

```python
# Plot Pole-Zero Map
ct.pzmap(sys, grid=True)
plt.title('Pole-Zero Map')
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
