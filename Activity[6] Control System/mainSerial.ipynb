```python
import serial 
import time

arduino = serial.Serial(port='COM5', baudrate=9600, timeout=.1)

time.sleep(2)

while True:
    command = input("Enter 1 to turn on the LED, 0 to turn it off, q to quit: ")

    if command == '1':
        arduino.write(b'1')
    elif command == '0':
        arduino.write(b'0')
    elif command == 'q':
        break
    else:
        print("Invalid command. Please enter 1, 0, or q.")

arduino.close()
```
